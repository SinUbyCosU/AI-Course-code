```python
# Define Neiborhood
def neighboor(state):
    row, col  =state
    print(f" row={row} and col= {col}")
    candidates= [("up",(row-1,col)), ('down',(row+1, col)), ("left", (row, col - 1)),("right", (row, col + 1))]
    print(f"candidates: {candidates}")
    result = []
    for action, (r, c) in candidates:
        print(f"r={r}, c={c}")
        if 0 <= r < height and 0 <= c < width and not walls[r][c]:
            result.append((action, (r, c)))
            print(f'action= {(action, (r, c))}')
    return result
    
```

### Stack frontier

```python
class StackFrontier():
    # Fuction for stack forntier
    frontier =[]
    def add(node):  #function to add elements in the Frontier 
        list.append(node)
        return(node)
    def remove(node):    #function to remove elements in the Frontier 
        global list
        node=list[-1]
        list=list[:-1]
        return node  
    def empty():
        return len(frontier)==0

    def contained_state(state):
        print(f"state is = {state}")
        for node in frontier:
            if node["state"]==state:
                return True
        return False      
    
```
